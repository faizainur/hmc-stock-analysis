# Predicting stock price

```python
dfreg = honda.loc[:,['Adj Close', 'Volume']]
dfreg['HL_PCT'] = (honda['High'] - honda['Low']) / honda['Close'] * 100.0
dfreg['PCT_Change'] = (honda['Close'] - honda['Open']) / honda['Open'] * 100.0

print(dfreg)
```

## Preprocess data

```python
import sklearn.preprocessing as preprocessing
dfreg.fillna(value=-99999, inplace=True)

forecast_out = int(math.ceil(0.01 * len(dfreg)))

forecast_col = 'Adj Close'
dfreg['label'] = dfreg[forecast_col].shift(-forecast_out)
X = np.array(dfreg.drop(['label'], 1))

X = preprocessing.scale(X)
8
X_lately = X[-forecast_out:]
X = X[:-forecast_out]

y = np.array(dfreg['label'])
y = y[:-forecast_out]
```

```python
clfreg = LinearRegression(n_jobs=-1)
clfreg.fit(X, y)

clfpoly2 = make_pipeline(PolynomialFeatures(2), Ridge())
clfpoly2.fit(X, y)

clfpoly3 = make_pipeline(PolynomialFeatures(3), Ridge())
clfpoly3.fit(X, y)
```

```python
def predict_stock(days, model):
    result = clfreg.predict(X[-days:])    
    return result
```

```python
linear_res = predict_stock(100, 'LinearRegression')
dfreg['Forecast'] = np.nan
print(linear_res)
```

```python
last_date = dfreg.iloc[-1].name
last_unix = last_date
next_unix = last_unix + datetime.timedelta(days=1)

for i in linear_res:
    next_date = next_unix
    next_unix += datetime.timedelta(days=1)
    dfreg.loc[next_date] = [np.nan for _ in range(len(dfreg.columns) - 1)]+[i]
```

```python
dfreg['Adj Close'].tail(250).plot(figsize=(30, 20))
dfreg['Forecast'].tail(100).plot(figsize=(30, 20))
plt.legend(loc=4, fontsize=20)
plt.title("HMC Stock Prediction", fontsize=22)
plt.ylabel("Price", fontsize=18)
plt.xlabel("Date", fontsize=18)
plt.show()
```
